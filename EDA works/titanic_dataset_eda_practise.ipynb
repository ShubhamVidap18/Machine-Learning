```python
plt.title('Null values in the Titanic Dataset')
df.isnull().sum().plot(kind = 'barh')
```

```python
# Checking the proportion of survivor for each sex

grouped = df.Survived.groupby(df.Sex)
sex_stats = pd.DataFrame({'total number':grouped.size(), 'number of survivors':grouped.sum()})
sex_stats.plot.bar()
```

```python
#distribution of age
df.Age.plot.hist(bins=50)
```

```python
#distribution of Fare
df.Fare.plot.hist()
```

```python
#distribution of Pclass
df.Pclass.plot.hist()
```

```python
age_group = pd.cut(df.Age, bins = 8)
age_grouped = df.Survived.groupby(age_group)
age_stats = pd.DataFrame({
    'total_number':age_grouped.size(),
    'number_of_survivor':age_grouped.sum()
    })
age_stats.plot.bar()

```

```python
#Checking if 'pclass' has any influence upon 'survive'

class_grouped = df.Survived.groupby(df.Pclass)
class_grouped_stats = pd.DataFrame({
    'total_number':class_grouped.size(),
    'numner_of_survivor':class_grouped.sum()
})

class_grouped_stats.plot.bar()
```
